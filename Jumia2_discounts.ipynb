```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.jumia.co.ke/?e=500&gad_source=1&gclid=Cj0KCQjwgL-3BhDnARIsAL6KZ6_jnMn2aIlJSiYcLMev2EHyDmi8jR0xzzPOOPRFZYnNxpnBj9kwG00aApjwEALw_wcB"

page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
div = soup.find('div', class_="row _no-g _6cl-4cm-shs -paxs")
articles = div.find_all('article')
```

```python
items = []
for article in articles:
    title = article.find('div', class_='name').text
    discount_tag = article.find('div', class_="bdg _dsct").text
    price_div = article.find('div', class_='prc')
    undisc_price = price_div.get('data-oprc')
    disc_price = article.find('div', class_='prc').text[4:]

    if discount_tag:
        items.append([title, discount_tag, undisc_price, disc_price])
```

```python
items
```

```python
df = pd.DataFrame(items, columns =['Title', 'Discount', 'Undiscounted_price', 'Discounted_price'] )
df.to_csv('Jumia2_discounts.csv', index=False)
```

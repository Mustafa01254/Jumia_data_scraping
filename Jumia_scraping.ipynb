```python
from bs4 import BeautifulSoup
import pandas as pd 
import requests
```

```python
url = 'https://www.jumia.co.ke/fashion-2-in-1-mens-short-sleeve-shorts-casual-set-khaki-161936451.html'
page = requests.get(url)

soup =  BeautifulSoup(page.content, "html.parser")
print(soup)
```

```python
title = soup.find('h1', class_='-fs20 -pts -pbxs').get_text().strip()
print(title)
```

```python
price = soup.find('span', class_='-b -ubpt -tal -fs24 -prxs').get_text().strip()[3:]
print(price)
    
    
    
```

```python
rating = soup.find('a', class_='-plxs _more').get_text().strip()[1:-1] 
print(rating)
```

```python
import datetime 
import time

today = datetime.date.today()
print(today)
```

```python
import csv
header = ['Title','Date', 'Rating', 'Price']

data = [title, today, rating, price]

with open('Jumia Tshirts.csv', 'w', newline='', encoding='utf-8')as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```

```python
def check_price():
    url = 'https://www.jumia.co.ke/fashion-2-in-1-mens-short-sleeve-shorts-casual-set-khaki-161936451.html'
    page = requests.get(url)

    soup =  BeautifulSoup(page.content, "html.parser")
    
    title = soup.find('h1', class_='-fs20 -pts -pbxs').get_text().strip()

    price = soup.find('span', class_='-b -ubpt -tal -fs24 -prxs').get_text().strip()[3:]

    rating = soup.find('a', class_='-plxs _more').get_text().strip()[1:-1]

    import datetime 
    import time

    today = datetime.date.today()
    import csv
    header = ['Title','Date', 'Rating', 'Price']

    data = [title, today, rating, price]

    with open('Jumia Tshirts.csv', 'a+', newline='', encoding='utf-8')as f:
        writer = csv.writer(f)
        writer.writerow(data)
```

```python
while(True):
    check_price()
    time.sleep(5)
```
